```python
seat_list=[]
seat_aspect_list=[]
seat_id_list=[]

df=pd.read_csv(r"C:/Users/Vaibhav/Downloads/FinalLabelsv2/FinalLabelsv2/Seat.csv")
df1=df[df['Aspect'].notna()]
df1=df1[df1['Seat'].notna()]
seat_list.extend(df1['Seat'])
seat_id_list.extend(df1['id_updated'])
#aspect_list=list(df1['Aspect'])
for asp in df1['Aspect']:
    seat_aspect_list.append(asp.lower())


r'''df2=pd.read_csv(r"C:/Users/Vaibhav/Documents/GitHub/Aspect-Based-Sentiment-Analysis/data/restaurants/SeatAspects.csv")
df3=df2[df2['Group'].notna()]
df3=df3[df3['Seat'].notna()]
seat_list.extend(df3['Seat'])
seat_id_list.extend(df3['id'])
#aspect_list=list(df1['Aspect'])
for asp in df3['Group']:
    seat_aspect_list.append(asp.lower())'''
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = list(stopwords.words('english'))
newStop = ['was', 'and', 'the', 'to', 'in', 'of', 'a', 'an', 'is', 'were', 'for', 'with', 'are', 'one', 'our', 'gave', 'have',
           'me', 'an', 'i', 'or', 'had', 'did', 'get', 'made', 'take', 'given', 'told', 'let', 'us','!']#, "n't"]
stop_words = stop_words+newStop
```

```python
vlsta=[]
vlstb=[]
vlstid=[]
for i in range(len(seat_list)):
    seat_list[i]=seat_list[i].replace("n't",' not')
    #vlst=[w for w in seat_list[i].split() and w not in stop_words]
    for word in seat_list[i].split():
            if word not in stop_words:
                vlsta.append(word)
                vlstb.append(seat_aspect_list[i])
                vlstid.append(seat_id_list[i])
    #for word in vlst#
    
df_seat_new=pd.DataFrame({'id':vlstid,'Word':vlsta,'Aspect':vlstb}).sort_values(by='Aspect')
```
